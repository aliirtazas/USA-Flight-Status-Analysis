```python
import pandas as pd
import glob

# Specify the path pattern for your Parquet files
parquet_files_pattern = '/content/*.parquet'

# Use glob to get a list of file paths matching the pattern
parquet_file_paths = glob.glob(parquet_files_pattern)

# Initialize an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Iterate through each Parquet file and append its content to the combined DataFrame
for file_path in parquet_file_paths:
    df = pd.read_parquet(file_path)
    combined_df = combined_df.append(df, ignore_index=True)

# Display the combined DataFrame
combined_df.desc
```
